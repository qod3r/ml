5. Вывести жанры игр с количеством игр, которые принадлежат соответствующему
жанру в виде "<жанр> - <количество игр>" для всех жанров, которые содержат в
своем названии не менее 3 различных гласных.

```python
def check_vowels(x):
    vowels = ['a', 'e', 'i', 'o', 'u']
    count = 0
    for v in vowels:
        if v in x:
            count += 1
    if count >= 3:
        return True
    return False
        
genres = df[df["Genre"].apply(check_vowels)]
res = {}
for genre in genres["Genre"].unique():
    g = genres[genres["Genre"] == genre]
    res[genre] = g['Name'].count()

res
```
